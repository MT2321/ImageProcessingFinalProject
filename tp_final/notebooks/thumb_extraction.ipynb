```python
import cv2 as cv
import matplotlib.pyplot as plt
import os
```

```python
root = "../../assets/videos"
img_root = "../../assets/imgs"
videos_path_raw = os.listdir("../../assets/videos")
videos_path = [root+"/"+path for path in videos_path_raw]
for video_path, video_name in zip(videos_path, videos_path_raw):
    print(video_path)
    video_capture = cv.VideoCapture(video_path)
    length_in_frames  = video_capture.get(cv.CAP_PROP_FRAME_COUNT)
    i = 0
    while i < length_in_frames//2:
        ret, frame = video_capture.read()
        i+=1
    ret, frame = video_capture.read()
    frame = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
    img_name = video_name.replace(".mp4", ".jpg")
    img_path =  img_root + "/" + img_name
    # plt.imshow(frame)
    print(img_path)
    plt.imsave(img_path, frame)
```
